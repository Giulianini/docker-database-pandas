```python
import pymongo
import pandas as pd
from pymongo import MongoClient
```

```python
connection_string = "mongodb://root:my_password@localhost:27017/?authMechanism=DEFAULT"
client = MongoClient(connection_string)
db = client["SampleCollections"]
collection_pokemon = db["samples_pokemon"]
df_pokemon = pd.DataFrame(list(collection_pokemon.find()))
```

```python
import matplotlib.pyplot as plt
import numpy as np
import PIL
import urllib


for idx, row in df_pokemon[["img", "name"]][0:3].iterrows():
    image = np.array(PIL.Image.open(urllib.request.urlopen(row["img"])))
    plt.suptitle(row["name"])
    plt.imshow(image)
    plt.show()
```
